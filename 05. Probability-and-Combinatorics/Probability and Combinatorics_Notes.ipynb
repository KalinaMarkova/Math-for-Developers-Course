# Probability and Combinatorics
### 1. Probability

The **scientific method** relies on experiments that use some **initial conditions** to provide an **outcome**. Usually, we control the initial conditions to isolate the outcome. It is used in descriptive statistics (to describe) or inferential statistics (to predict).

A **random event** is a set of outcomes of an experiment. Each outcome happens with a certain **probability**.

A **random variable** is an expression whose value is the outcome of the experiment. They are usually denoted with capital letters (X, Y, Z).

It is not possible to predict the next outcome of a random event. However, with more **trials** patterns and laws become more apparent.

A series of trials under the same conditions constitutes a **controlled experiment**.

The formula for experimental frequency (f) is as follows:

$$ f_i = \frac {m_i}{n} $$

m = number of trials that resulted in the specific outcome
n = all trials
i = possible outcomes

As n increases, $f_i $ "stabilizes" around some number. As we cannnot perform infinitely many experiments, we can "extend" the trials mathematically using the concept of limits to get the **probability outcome** of A:

$$ p(A) = \lim _{n \to  \infty} \frac{m}{n} $$

The value we always get for **p** is always **between 0 and 1**. It is also a theoretical value that can be approximated very precisely.

The sum of all probabilities of a controlled experiment is equal to 1. If it is not, there must be a mistake somewhere.

**Measurement** is the process whereby a feature is evaluated. In any given study, you are trying to measure or evaluate certain **elements that change value**, depending on certain factors. This is the definition of **variables**. For example, height is a variable as it changes from person to person. Other variables change over time, e.g. the level of anyone's anxiety. A person's age is another example of a variable that changes over time. 

There are three main types of variables: 
* **continuous variables** can take any numerical value and are measured;
* **discrete variables** can only take certain numerical values and are counted, e.g. a dice roll, as it has a certain number of outcomes, and nothing else, we can roll a 4 or a 5, but not a 4.6;
* **categorical variables** involve non-numeric groups or categories (e.g., hair color, gum flavor, dog breed, and cloud type).

**Note:** For our purposes, we will use almost interchangeably** discrete and categorical variables as they are similar.

A **probability densitty** is the probability of the result being in a tiny interval $dx $:

$$ dp = \frac{dx}{b-a} $$

where $a, b $ are both ends of the interval.

### Combinatorics

**Combinatorics** provides a way to count things, especially a large number of things, without counting them explicitly. For example, if you want to count explicitly every possible move in chess, it will take you many more times the life of the universe.

The prerequisites to use combinatorics are:
* finite (countable) number of outcomes;
* all outcomes have equal probabilities.

Such examples are roulette and card games. Or the number of ways a deck of cards can be shuffled, which is $ 52! $ or $ 54! $ if the joker cards are included.

All outcomes are denoted with $n $, while all experiment outcomes are presented with $k $. Usually, $n $ is fixed, while $k $ depends on the experiment.

Types of samples:
* with repetition / without repetition
* ordered / unordered

**Rule of the sum:** If $m $ and $n $ are choices for different actions that cannot be done at the same time, then there are $m $ + $n $ ways to choose one of these actions.

**Rule of the product:** If the two actions are performed one after the other, there $m $.$n $ ways to do both actions (if both choices are independent).

A **tree diagram** is a graphic method of solving combinatorial problems.



A **permutation** ($ P_n$) is shuffling of elements without repetition. If A = {1, 2, 3, 4}, then some permutations are {1, 4, 3, 2}, {2, 3, 4, 1}. There are $n $ choices for the first element, $ n - 1 $ for the second element, $ n - 2 $ for the third and so on. For the last one, there is only one choice. Total:

$$ n! = 1.2.3\dots.n$$



A **variation** ($V_n^k $) is a**n ordered subs**et o$f $ elements from $A$. "Variation of $n$ elements, $k^th$ class". If A = {1, 2, 3, 4}, then some variations are {1,2}, {4,3}, {3,1}. There are $n $ choices for the first element, $ n - 1 $ for the second element and so on. For the last one, there are $ n - k + 1$ choices. Total:

$$ V_n^k = n.(n - 1).\dots.(n - k + 1) = \frac{n!}{(n - k)!} $$

A **combination** ($C_n^k $) is an **unordered subset** of $k $ elements from $A$. If A = {1, 2, 3, 4} and $k = 2$, then some combinations are {1,2}, {4,3}, {3,1}.

$$ C_n^k = \frac{n!}{(n-k)!k!} $$

This is also known as "$n$ choose $k$".

$$ \left(\frac{n}{k}\right) = \frac{n!}{(n-k)!k!} $$


### Probability Algebra

An **event** is a result of the experiment. An **elementary event** is one particular outcome. A **compound event** consists of many elementary events. For example, getting an odd number from a dice. An **event space** is the set $\Omega$ of all possible events.

* If an event A happens with event B, A is a **consequence** of B: $ A \subset B$
* If $ A \subset B$ and $ B \subset A$, then $ A = B$
* **Complementary event:** $\overline{\rm A}$ happens if $A$ does **not** happen
* **Impossible event:** contains no elementary events: $\emptyset$
* **Product of events**: happens if **A and B** happen: $ C = A \cap B$
  * **Incompatible events:** $ A \cap B = \emptyset$
* **Sum of events:** happens if A, B or both happen:  $ C = A \cup B$
  * If A and B are incompatible, $C = A + B$
    
Logical relations are the same as set operations and event operations:
* **AND**: intersection (in this intersection both events have happened)
* **OR**: union
* **NOT**: compliment

**Conditional probability** - additional information about the experiment outcome can change the probabilities. 

**a priori $\rightarrow$ a posteriori**

The probability of event A given that event B has happened is notated as: $P(A|B)$
In this case, the probability $P(A|B)$ corresponds to the part of the set B which is shared with A.

$$ P(A|B) = \frac{P(A \cap B)}{P(B)} $$



If the events are independent, the information about B, the event that has happened, does not bring any information about A. In the case of two independent events, their probability is multiplied:

$$ P(A \cap B) = P(A).P(B) $$

The **Bayes' theorem** is the conditional probability of an event A, given the occurrence of another event B. It tells how to update the probabilities when we know some evidence:

$$ P(A|B) = \frac{P(B|A)P(A)}{P(B)} $$

It is used in spam detection.

The Monty Hall problem - do not trust your intuition about statistics, math always wins.

#### Statistical Distributions

Instead of graphing the real function, we can make a histogram of counts. This gives us a much better idea what the random variable looks like, for example frequency distribution.

A **discreet distribution** is represented by a table that maps each outcome of a random variable to a probability. A histogram of every single value of a discreet distribution is called a **probability mass function**.

$$p_X(x_i) = P(X = x_i) $$

A **cumulative distribution function** is represented by a table that maps each outcome of a variable to the probability of its value being less or equal to a given number:

$$ F_X(x_i) = P (X \le x_i) $$

It is also called **cumulative mass function**. Every such function is non-decreasing (also called  monotonic increasing function), it usually starts at 0 and always ends at 1.

A **continuous distribution**  describes the probabilities of the possible values of a continuous random variable. A continuous random variable is a random variable with a set of possible values (known as the range) that is infinite and uncountable. In this case, it is called **cumulative density function**. Probabilities of continuous random variables (X) are defined as the area under the curve of its probability density function. Thus, only ranges of values can have a nonzero probability. The probability that a continuous random variable equals some value is always zero.

#### Common Distributions

The **Bernoulli distribution** is the discrete probability distribution of a random variable which takes the value 1 with probability $p$ and the value 0 with probability $q = 1 - p$. Normally, it can be thought of as a model for the set of possible outcomes of any single experiment that asks a yes-no question. Such questions lead to outcomes that are like Boolean values. Example: a flip of a fair coin.

In the **uniform distribution** all values in some range are equally likely.

The **normal (Gaussian) distribution** is a type of continuous probability distribution for real-valued random variable. Its probability density function is:

$$f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e^{ -\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2 }$$



In the **standard normal distribution** $\mu = 0 $ and $\sigma = 1$. Every normal distribution can be converted to a standard normal distribution. To do this, from each value we need to subtract the $\mu$ and divide by the $\sigma$. This operation is called **Z-score**. In this way, we have the same frame for different things, also referred to as **standardization**.

When there are many independent sources of errors, the sum of all errors tends to be close to zero. And smaller mistakes are much more probable than large mistakes. The sum of many independent random variables tends to a normal distribution even if the original variables are not normally distributed. This is the case if we have large enough sample. $n \in [30; 40]$
